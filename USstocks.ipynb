```python
import warnings
import pandas as pd
import numpy as np

usstocks = pd.read_csv("data/SPY.csv")
usstocks["Date"] = pd.to_datetime(usstocks["Date"])
usstocks = usstocks.set_index(["Date"])
usstocks = usstocks.sort_index()
usval = usstocks["Adj Close"]
```

#### <font color=SteelBlue>Following statement to import Class</font>

```python
from CATSpredict import CATS
```

#### <font color=SteelBlue>Without Progress Bar Display</font>
*In the case where tqdm library (progress bar library) is not available.*

```python
cat = CATS(usval)
```

*Check analysis of TimeSeries for stationarity, trend-stationarity and presence of seasonality. If seasonality present, use seasonal models.*

```python
cat.analysis(season=[60, 480])
```

<br>

*<b>predict_next_auto()</b> method would search for the best forecasting model and return the one step prediction interval of the time series.*

```python
cat.predict_next_auto()
```

#### <font color=SteelBlue>With Technical Labels</font>
*If detail=True, prediction intervals generated will be labelled by their respective technical terms. Default: detail=False.*

```python
cat = CATS(usval, pbar=True)
```

```python
cat.predict_next_auto(detail=True)
```

*<b>predict_next_auto()</b> method would search for the best forecasting model and return the one step prediction interval of the time series. While searching for the bestmodel, the method will save a summary of the best parameters and effectiveness of the models in either **nonseasonal_result** or **seasonal result attribute** depending on seasonal parameter of the method.*

```python
cat.nonseasonal_result
```

*<b>non_seasonal()</b> method would search for the best forecasting model and return a summary of the best parameters and effectiveness of the models, while saving the summary in either **nonseasonal_result** or **seasonal result attribute**. Results found in search process will be stored in respective attributes. Similar for <b>seasonal() method.</b>*

```python
cat.non_seasonal()
```

```python
cat.residiid
```

```python
cat.engle
```

```python
cat.standardiid
```

```python
cat.standard_engle
```

```python
cat.bestmodel
```

*After **non_seasonal()** or **seasonal()** has been run, **bestmodel** attribute will be found and **predict_with_bestmodel()** method will forecast the next immediate prediciton interval using **bestmodel**.*

```python
cat.predict_with_bestmodel()
```

*Prediction interval can also be generated manually by inputting model and parameters using **predict_next** method.*

```python
cat.predict_next((2,2), order=(3,2,4))
```
